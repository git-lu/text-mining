# Scrapping obras

```python
url_artista = 'https://coleccion.malba.org.ar/artistas/portinari-candido/'
```

```python
url_artista_una_obra = 'https://coleccion.malba.org.ar/artistas/accinelli-pablo/'
```

```python
def scrap_links_obras(soup):
    links = []
    tables = soup.findAll('div', attrs = {'class':"col-md-3 col-sm-6 item"})
    for row in tables:
        links.append(row.a['href'])    
    return links

def scrap_ficha_tecnica(soup):
    ft = soup.find('div',attrs={'id':'side-info'})
    ficha = ft.text.strip().split("\n")
    ficha_tecnica = [f.strip() for f in ficha ]
    return " - ".join(ficha_tecnica)

def scrap_textos(soup):
    textos = []
    table = soup.findAll('div', attrs = {'id':"textos"})     
    for texto in table:
        texto = texto.text.strip()
        textos.append(texto)
    return "".join(textos)

def get_obra_info(link_obra,artista_url):
    obra = {}
    obra_soup = get_soup(link_obra)
    obra['ficha_tecnica'] = scrap_ficha_tecnica(obra_soup)
    obra['link_obra'] = link_obra
    obra['texto'] = scrap_textos(obra_soup)
    obra['link_autor'] = artista_url
    return obra

def get_obras_por_artista(artista_url):
    obras = []
    soup_artista = get_soup(artista_url)
    links_obras = scrap_links_obras(soup_artista)
    if links_obras:
        for link in links_obras:
            obra = get_obra_info(link,artista_url)
            obras.append(obra)
    else:
        obra = get_obra_info(artista_url,artista_url)
        obras.append(obra)
    return obras
```
